#Steps
- Generate N solutions
- Generate m solutions
  - choose parent
  - crossover + mutation
- Generate N + m solutions
- Repeat from step 2

```python
W = 20
items = 7
v = [50,24,33,50,32,1,36]
w = [10,5,7,8,7,2,8]
```

```python
#N Solution
import random

def generate_population(n, proba):

  more_solutions = []
  for _ in range(n):
    solution = []
    for i in range(len(w)):
      if random.random() < proba:
        solution.append(1)
      else:
        solution.append(0)
    more_solutions.append(solution)
  return more_solutions

#generate_population(100,.5)
```

```python
#Generate m child solutions

def roulette_selection(n, m):
  #equal proba
  proba = .5
  random_selection = random.choices(m, k = n)
  
  return random_selection



def crossover(s1,s2, split = None):
  if split == None:
    split = random.randint(1,len(s1)-1)
  #print(f'this is the index of the split: {split}')
  
  #print(f'this is s1: {s1}')
  #print(f'this is s2: {s2}')
  
  child_s1 = s1[:split] + s2[split:]
  child_s2 = s2[:split] + s1[split:]
  #print(f'this is child_1: {child_1}')
  #print(f'this is child_2: {child_2}')
  return child_s1, child_s2

s1,s2 = roulette_selection(2, generate_population(10,.5))[0], roulette_selection(2, generate_population(10,.5))[1]

p1 = [0,0,0,0,0,0,0]
p2 = [1,1,1,1,1,1,1]
crossover(s1,s2)
```

```python
#Mutation

def mutation(child,proba):

  for i in range(len(child)):

    if random.random() < proba:
      if child[i] == 0:
        child[i] = 1
      else:
        child[i] = 0

  return child

child_1,child_2 = crossover(s1,s2)[0], crossover(s1,s2) [1]
m_1,m_2 = mutation(child_1, .5), mutation(child_2,.5)
m_1,m_2
```

```python
#pass in population, pass in the number you need to keep
initial_pop = generate_population(10, .5)

def reproduction(n):
  new_pop = []

  for _ in range(len(n)):
    s1,s2 = roulette_selection(2, n)[0], roulette_selection(2, n)[1]
    crossover_selection = crossover(s1,s2)
    mutation_selection = mutation(crossover_selection[0],.5), mutation(crossover_selection[1],.5)

    new_pop.append(mutation_selection)
  
  return new_pop

#reproduction(initial_pop)
```

```python

def population_m_n(pop):
  pop = reproduction(initial_pop)
  pop_list = []
  for i in range(len(pop)):
    for j in range(0,2):
      pop_list.append(pop[i][j])
  new_pop = pop_list + initial_pop
  return new_pop

#population_m_n(initial_pop)
```

```python

def fitness_check(pop,v=v,w=w,W=W):
  value_list = []
  weight_list = []
  for i in range(len(pop)):
    total_value = 0
    max_weight = 0
    for j in range(0,7):
      if pop[i][j] == 1:
        #print(pop[j])
        total_value += v[j]
        max_weight += w[j]
    if max_weight > W:
      total_value = total_value * -1
    value_list.append(total_value)
    weight_list.append(max_weight)
  total_list = zip(pop,weight_list,value_list) 
  return zip(pop,weight_list,value_list)
```

```python
def fitness_check_v1(v1,v=v,w=w,W=W):
  total_value = 0
  max_weight = 0
  #print(v)

  for i in range(len(v1)):
    if v1[i] == 1:
        #print(v1[i])
        total_value += v[i]
        max_weight += w[i]
    if max_weight > W:
      total_value = total_value * -1
  return total_value

#fitness_check_v1(population[0])
```

```python
population = population_m_n(initial_pop)

def tournament_survival(p,n):
  mega_list = []
  for _ in range(n):
    survived_values = []
    survived_solutions = []
    for i in range(len(p)):
      if fitness_check_v1(p[i]) > fitness_check_v1(p[random.randint(1, len(p)-1)]):
        survived_solutions.append(p[i])    
        survived_values.append(fitness_check_v1(p[i]))
    mega_list = zip(survived_solutions,survived_values)
  #print(survived_values)
  
  return mega_list

final=list(tournament_survival(population,100000))
final
```
