```python
import random as rand
import math

class City:
  def __init__(self,name):
    self.name = name
    self.x = rand.random()
    self.y = rand.random()


cities = []
for i in range(6):
  cities.append(City(i))

def distance(a,b):

  return math.sqrt((a.x - b.x)**2 + (a.y - b.y)**2)


```
