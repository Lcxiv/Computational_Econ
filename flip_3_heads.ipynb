```python
import random

def flip(n=5):
  count = 0
  for _ in range(n):
    if random.randint(0,1) == 1:
      count+=1

  if count == 3:
    return 1

  return 0

def montecarlo(n):
  results = []

  for i in range(n):
    results.append(flip())

  return print(f'The probability of drawing three heads in after 5 flips is: {sum(results)/len(results)} after {n} simulations!')

for i in range(1,100, 10):  
  montecarlo(i)

for i in range(100,10000, 1000):  
  montecarlo(i)
```
