```python
import random as rand 

#variables

initial_cash = 1000
bet = 10 
trials = 10000
success = 0
winnings = 0
bets = []

def bet_red(bet):
  draw = rand.randint(0,37)
  if draw != 0 and draw % 2 == 0:
    return bet * 2
  return 0



for _ in range(trials):
  bet_count = 0

  remaining_cash = initial_cash

  while remaining_cash > 0 and remaining_cash < 2 * initial_cash :
    #print(remaining_cash)
    remaining_cash -= bet
    remaining_cash += bet_red(bet)
    bet_count +=1
    #print(remaining_cash)
  success += remaining_cash > 0
  winnings += remaining_cash
  bets.append(bet_count)


print(f'we won {success/trials}, and on average: {winnings/trials}')
print(f'avg number of bets {sum(bets)/trials}')
print(f'max bets {max(bets)} and {min(bets)}')

```
