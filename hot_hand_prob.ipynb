```python
import random 

coin = ['heads', 'tails']

def flip():
  serie = []
  for _ in range(4):
    serie.append(random.choices(coin))

  return serie

flip()

results = flip()
count = 0
head_count = 0

#print(results)
freq = []
for _ in range(10):
  for i in results:
    #print(i)
    if i == coin[0]:
      print('true')
      if i+1 == coin[0] or i-1 == coin[0]:
        head_count += 1
        count += 1
        relative_freq = head_count/count
        freq.append(relative_freq)
    else:
      count += 1 
#print(f'the number of times having two heads following each other is: {head_count/count}')

flips = 4
trials = 100

frequency = []
second_heads_count = 0 
for _ in range(trials):
  flip_results = random.choices(['H','T'], k = flips)
  second_heads_count += 1
  prev = 'T'
  for result in flip_results:
    if result == 'H' and prev == 'H':
      second_heads_count += 1
    prev = result
first_heads_count = sum([f == 'H' for f in flip_results[0:-1]])

if first_heads_count > 0:
  frequency.append(second_heads_count / first_heads_count)
print(f'Average Frequency: {sum(frequency) / len(frequency)}')
```
